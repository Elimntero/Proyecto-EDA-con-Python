```python
# Importamos pandas.
import pandas as pd
```

```python
# Leemos csv y creamos el primer dataset de clientes.
bank_dataset = pd.read_csv('bank-additional.csv', sep=',')
bank_dataset
```

```python
# Mostrar hojas disponibles en el Excel
print("Hojas del Excel:", 'customer-details.xlsx')
```

```python
# Cargar las tres hojas del Excel
customer_2012 = pd.read_excel('customer-details.xlsx', sheet_name="2012")
customer_2013 = pd.read_excel('customer-details.xlsx', sheet_name="2013")
customer_2014 = pd.read_excel('customer-details.xlsx', sheet_name="2014")
```

```python
# Unimos todos los datos de los clientes en un solo DataFrame
customer_df = pd.concat([customer_2012, customer_2013, customer_2014], ignore_index=True)
customer_df
```

```python
# Cambiamos en bank-additional el nombre de la columna 'id_' y en customer-details el nombre de la columna 'ID' ambas a 'customer_id'
bank_dataset.rename(columns={'id_': 'customer_id'}, inplace=True)
customer_df.rename(columns={'ID': 'customer_id'}, inplace=True)
```

```python
bank_dataset['customer_id'] = bank_dataset['customer_id'].astype(str)
customer_df['customer_id'] = customer_df['customer_id'].astype(str)
```

```python
# Tras los pasos anteriores ya podemos unir los dos dataset usando la columna en común 'customer_id'
merged_df = pd.merge(bank_dataset, customer_df, on='customer_id', how='inner')
merged_df
```

```python
# Comenzamos a realizar el tratamiento de datos de nuestro conjunto completo de datos 

# Empezamos por utilizar la funcion info(), que en pandas resulta útil para obtener un resumen del conjunto de datos. 
merged_df.info()
```

```python
# Eliminamos duplicados
merged_df.duplicated().any()
```

En este caso no hay datos duplicados, por lo que no se elimina ninguno.

```python
# Revisamos los posibles valores NaN
print("Cantidad de nulos de nuestro conjunto de datos de Bank y Customer")

print(merged_df.isna().sum(axis=0))
```

```python
# Existen columnas que sus valores son de tipo object y reflejan las opciones Sí o No, 
# en nuestra lista de datos/columnas vemos que contienen 0 (refleja la opcion "NO") y 1 (refleja o referencia la opcion "SI")

columnas_0_1 = ['default', 'housing', 'loan', 'y']

# Reemplazar 0 por 'No' y 1 por 'Sí' en las columnas especificadas
merged_df [columnas_0_1] = merged_df[columnas_0_1].replace({0: 'No', 1: 'Sí'})
merged_df
```

```python
# Personalmente pienso que, en los casos de las propiedas'Unnamed: 0_y' y 'Unnamed: 0_x' no serían necesarias en nuestro conjunto de datos, 
# al igual que las columnas 'Unnamed: 24', 'Unnamed: 25' y 'Unnamed: 26' que no estan definidas en el documento especifico del proyecto. 
# Si estuvieramos hablando de datos generales, con saber el ID como dato primario sería suficiente.

merged_df = merged_df.drop(columns=['Unnamed: 0_x','Unnamed: 0_y'], axis=1)
merged_df
```

```python
#Pasamos la columna "marital" de mayúscula a minúscula.
merged_df ['marital'] = merged_df ['marital'].str.lower()
merged_df
```

```python
# 1. Estadísticas generales.
print("Estadísticas generales")
print(merged_df.describe())
```

```python
# 2. Distribución por edad, género, estado civil, nivel educativo, etc.
# ¿Quién aceptó la campaña según edad, género, educación?
print("\nAceptación por nivel educativo:")
print(merged_df.groupby('education')['y'].value_counts(normalize=True))

print("\nAceptación por estado civil:")
print(merged_df.groupby('marital')['y'].value_counts(normalize=True))
```

```python
# Edad promedio de quienes aceptaron
print("\nEdad promedio por aceptación:")
print(merged_df.groupby('y')['age'].mean())
```

```python
# 3. Comportamiento de compra y relación con variables del banco.bank_dataset.bank_dataset.
# Análisis de correlación entre variables numéricas.
correlation_matrix = merged_df.select_dtypes(include='number').corr()
print(correlation_matrix)
```

+1.0: correlación positiva perfecta → cuando una variable sube, la otra también.
-1.0: correlación negativa perfecta → cuando una sube, la otra baja.
0.0: sin correlación lineal.

```python
# Mapa de calor: Correlación entre variables numéricas.
corr = merged_df.select_dtypes(include='number').corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Matriz de correlación")
plt.show()
```

Los valores van de -1 a 1:
+1: correlación positiva perfecta (ambas variables suben o bajan juntas).
-1: correlación negativa perfecta (una sube y la otra baja).
0: sin correlación lineal.

Los colores representan la fuerza de la correlación:
· Rojo fuerte: correlación negativa fuerte.
· Azul fuerte: correlación positiva fuerte.
· Tonos suaves o blancos: correlación débil o nula.

Cuanto más días pasaron desde el último contacto, menos contactos previos tuvo el cliente.
campaign y emp.var.rate: correlación positiva débil (~0.15)
Leve relación entre campañas más insistentes y la tasa de empleo.
En general, la mayoría de las variables tienen correlaciones débiles o nulas entre sí.
Esto sugiere que las variables son en su mayoría independientes, lo cual es bueno para modelos predictivos porque evita redundancia.
